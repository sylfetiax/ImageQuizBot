```python
import os
import shutil

# Шлях до вашої вихідної структури папок
source_path = "data"

# Шлях до нової структури папок
target_path = "data_new"

# Створення нової структури папок
new_structure = ['triangle', 'ellipse', 'rectangle', 'other']

for folder in new_structure:
    os.makedirs(os.path.join(target_path, folder), exist_ok=True)

# Обхід старої структури та переміщення файлів
for root, dirs, files in os.walk(source_path):
    for file in files:
        # Отримання шляху до поточного файлу та ім'я папки
        file_path = os.path.join(root, file)
        folder_name = os.path.basename(os.path.dirname(file_path))

        # Визначення категорії файлу на підставі назви папки
        if folder_name.lower() == 'triangle':
            category = 'triangle'
        elif folder_name.lower() == 'ellipse':
            category = 'ellipse'
        elif folder_name.lower() == 'rectangle':
            category = 'rectangle'
        else:
            category = 'other'

        # Визначення нового шляху для переміщення файлу
        new_path = os.path.join(target_path, category, file)

        # Копіювання файлу разом з атрибутами
        shutil.copy2(file_path, new_path)

# Шлях до директорії, в якій розташовані папки з файлами
directory_path = "data_new"

# Обхід усіх папок у заданій директорії
for folder_name in os.listdir(directory_path):
    folder_path = os.path.join(directory_path, folder_name)

    # Перевірка, чи є поточний шлях директорії папкою
    if os.path.isdir(folder_path):
        # Обхід файлів у папці та видалення файлів з розширенням .csv
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)

            # Перевірка, чи є поточний шлях файлом та чи має розширення .csv
            if os.path.isfile(file_path) and file_name.lower().endswith('.csv'):
                # Видалення файлу
                os.remove(file_path)
                print(f"Файл {file_path} видалено.")
```

```python
directory_path = "data_new/val"

# Обхід усіх папок у заданій директорії
for folder_name in os.listdir(directory_path):
    folder_path = os.path.join(directory_path, folder_name)

    # Перевірка, чи є поточний шлях директорії папкою
    if os.path.isdir(folder_path):
        # Отримання списку файлів у папці
        files_in_folder = os.listdir(folder_path)

        # Підрахунок кількості файлів у папці
        num_files = len(files_in_folder)

        print(f"У папці '{folder_name}' знаходиться {num_files} файл(ів).")
```

```python
import random

# Шлях до основної директорії з категоріями зображень
base_path = "data_new"

# Папка для переміщення вибраних зображень
val_folder = os.path.join(base_path, "val")
os.makedirs(val_folder, exist_ok=True)

# Категорії зображень
categories = ['triangle', 'ellipse', 'rectangle', 'other']

# Відсоток зображень, які вибрати
percentage_to_select = 7

# Обхід кожної категорії та вибір 15% зображень
for category in categories:
    category_path = os.path.join(base_path, category)
    val_category_path = os.path.join(val_folder, category)
    os.makedirs(val_category_path, exist_ok=True)

    # Отримання списку всіх файлів у категорії
    all_files = os.listdir(category_path)

    # Визначення кількості файлів для вибору (15%)
    num_files_to_select = int(len(all_files) * (percentage_to_select / 100))

    # Випадковий вибір 15% зображень
    selected_files = random.sample(all_files, num_files_to_select)

    # Переміщення вибраних файлів у папку "val"
    for file_name in selected_files:
        source_path = os.path.join(category_path, file_name)
        target_path = os.path.join(val_category_path, file_name)
        shutil.move(source_path, target_path)
        print(f"Файл {file_name} переміщено до папки 'val/{category}'.")

print("Операція завершена.")
```
